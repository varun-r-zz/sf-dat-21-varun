Answer: Using the Admissions dataset from a UCLA Graduate Program, we went to understand what effect undergraduate GPE, GRE Score and Alma Mater prestige have on admissions. Ideally we would like to understand what factors are most important in predicting this. Due to the Predicted Variable being categorical, we should use a classification model

```python
import pandas as pd

df = pd.read_csv('../../Dataset/admissions.csv')
df = df.dropna()


dummy_ranks = pd.get_dummies(df.prestige, prefix = 'prestige').astype(int)
df = df.join([dummy_ranks])

df.describe()
df.groupby('admit').describe()
```

```python
pd.crosstab(df.prestige, df.admit, margins=True)
```
